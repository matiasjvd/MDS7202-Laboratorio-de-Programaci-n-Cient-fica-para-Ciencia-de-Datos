Describir qué métricas se ocupan para la evaluación del problema y por qué se utilizan.

Para evaluar el desempeño de los modelos, tanto para el problema de clasificación como el de regresión se utilizarán dos métricas, una para cada problema, qué nos indicarán cuál es el grado de acierto de las predicciones generadas.

##### Métrica para el problema de clasificación:

Para los problemas de clasificación existen métricas muy comúnmente utilizadas, llamadas precision y recall. Sobre cada categoría a clasificar (en nuestro caso  'Negative', 'Mixed', 'Mostly Positive', 'Positive' y 'Very Positive') se puede calcular la métrica de precision y recall, éstas corresponden a la porción de predicciones correctas de la categoría sobre el total de veces que el clasificador predijo dicha categoría, en el caso de la métrica **precision**; y la porción de predicciones correctas de la categoría sobre el total de elementos que realmente correspondía a la categoría, en el caso del **recall**.  A partir de estas dos métricas, se define una tercera métrica llamada f1_score, que se calcula para cada categría como un promedio armónico de ambas, bajo la fórmula siguiente:

$$ F1_{i} = 2 * \frac{precision_i \cdot recall_i}{precision_i + recall_i}$$

Donde $precision_i$ , $recall_i$ y $F1_i$ corresponden al precision, recall y F1 de la categoría i.

Finalmente, para calcular un F1 Score general del clasificador (y no para cada categoría) calculamos un promedio ponderado de los f1 en cada categoría, donde las ponderaciones son proporcionales al tamaño de la muestra de cada categoría (por ejemplo, en un dataset perfectamente balanceado, el f1_wighted corresponde a un promedio simple de los f1_score para cada categoría).

**Por qué se utiliza esta métrica?**

Esta métrica mide el rendimiento general del clasificador, y al ponderarlo por cada tamaño de muestra permite un resultado acorde a los datos inputados al clasificador. Generalmente, en otros problemas de clasificación son utilizadas las métricas precission o recall (que también tienen su versión *weighted*) cuando se trabaja con datos sensibles en los cuales es muy importante que el clasificador tenga cierto tipo de error muy bajo y no es tan importante otros tipos de error (por ejemplo, si buscamos clasificar imágenes de tumores en benignos y malignos, quisiéramos reducir al mínimo las posibilidades de que el clasificador prediga mal tumores malignos); como éste no es el caso, la métrica *f1_weighted* es la más apropiada ya que conjuga cada tipo de error posible del clasificador y nos entrega una visión más general de su desempeño.

##### Métrica para el problema de regresión:

Para los problemas de regresión una de las métricas más utilizadas es el $R^{2}$ score. Esta métrica nos indica qué tan buenas son nuestras predicciones de regresión, valiendo 1 en el caso de predicción sin error y valiendo 0 para una predicción constante igual al promedio de la variable objetivo, sin importar el valor de las características (siempre que la variable objetivo no sea constante).

Esta métrica se calcula bajo la siguiente fórmula:

$$ R^{2} = 1 - \frac{\sum_{i = 1}^{n} (y_{i}-\hat{y}_{i})^2}{\sum_{i = 1}^{n} (y_{i}-\bar{y})^2} $$

Donde $y_{i}$ es la observación i-ésima de la variable a predicir y, $\hat{y}_{i}$ es la predicción i-ésima de la variable y (generada por el modelo) y $\bar{y}$ corresponde al promedio de la variable objetivo.

Así, esta métrica compara el desempeño de nuestras predicciones frente a las predicciones hechas por un regresor *dummy* (regresor constante) igual al promedio de la variable objetivo.

Esta métrica es atingente al problema ya que marca como baseline un regresor dummy igual al promedio. Así, consideramos un modelo como un buen predictor cuando supera largamente este tipo de regresores (lo cual es muy razonable). Generalmente, existen otras métricas para problemas de clasificación (como MAE, MSE, MAPE o RMSE) pero este tipo de métricas son útiles cuando tenemos un modelo base que buscamos mejorar, cuando no se tiene la métrica $R^2$ permite compararlo con un modelo base sencillo.

```python
#Sabemos que las variables achievements y average_playtime tienen outliers, pero no los eliminaresmos, pero mas adelante los normalizaremos

#Eliminaresmos los outliers de las variables achievements y average_playtime, creando una funcion para ello
def remove_outliers(df_copy, column):
    q1 = df_copy[column].quantile(0.25)
    q3 = df_copy[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    df_copy = df_copy[(df_copy[column] > lower_bound) & (df_copy[column] < upper_bound)]
    return df_copy

df_train = remove_outliers(df_train, 'achievements')
df_train = remove_outliers(df_train, 'average_playtime')

```

```python

stop_words = stopwords.words('english')
# Definimos un tokenizador con Stemming
class StemmerTokenizer:
    def __init__(self):
        self.ps = PorterStemmer()
    def __call__(self, doc):
        doc_tok = word_tokenize(doc)
        doc_tok = [t for t in doc_tok if t not in stop_words]
        return [self.ps.stem(t) for t in doc_tok]

# Inicializamos tokenizador
tokenizador = StemmerTokenizer()



bog = CountVectorizer(tokenizer= tokenizador)
```

### Anexo: Generación de Archivo Submit de la Competencia

Para subir los resultados obtenidos a la pagina de CodaLab utilice la función `generateFiles` entregada mas abajo. Esto es debido a que usted deberá generar archivos que respeten extrictamente el formato de CodaLab, de lo contario los resultados no se veran reflejados en la pagina de la competencia.

Para los resultados obtenidos en su modelo de clasificación y regresión, estos serán guardados en un archivo zip que contenga los archivos `predicctions_clf.txt` para la clasificación y `predicctions_rgr.clf` para la regresión. Los resultados, como se comento antes, deberan ser obtenidos en base al dataset `test.pickle` y en cada una de las lineas deberan presentar las predicciones realizadas.

Ejemplos de archivos:

- [ ] `predicctions_clf.txt`

        Mostly Positive
        Mostly Positive
        Negative
        Positive
        Negative
        Positive
        ...

- [ ] `predicctions_rgr.txt`

        16103.58
        16103.58
        16041.89
        9328.62
        107976.03
        194374.08
        ...



```python
from zipfile import ZipFile
import os

def generateFiles(predict_data, clf_pipe, rgr_pipe):
    """Genera los archivos a subir en CodaLab

    Input
    predict_data: Dataframe con los datos de entrada a predecir
    clf_pipe: pipeline del clf
    rgr_pipe: pipeline del rgr

    Ouput
    archivo de txt
    """
    y_pred_clf = clf_pipe.predict(predict_data)
    y_pred_rgr = rgr_pipe.predict(predict_data)
    
    with open('./predictions_clf.txt', 'w') as f:
        for item in y_pred_clf:
            f.write("%s/n" % item)

    with open('./predictions_rgr.txt', 'w') as f:
        for item in y_pred_rgr:
            f.write("%s/n" % item)

    with ZipFile('predictions.zip', 'w') as zipObj2:
       zipObj2.write('predictions_rgr.txt')
       zipObj2.write('predictions_clf.txt')

    os.remove("predictions_rgr.txt")
    os.remove("predictions_clf.txt")
```
