````python
other_grades = pd.read_csv("C:/Users/matia/OneDrive/Escritorio/Material U 2022-2/Laboratorio de Programaci칩n Cient칤fica para Ciencia de Datos/Laboratorios/Lab 4/other_grades.csv")
other_grades.head()
````

````python
#Outer join
pd.merge(
    left=df_grades,
    right=other_grades,
    on="names",
    sort=True,
    how="outer",
)
#Podemos ver que haciendo un outer join, se unen los dos dataframes, pero se crean valores nulos en las columnas que no tienen datos en com칰n, 
#ademas de que se repiten los nombres de las personas que estan en ambos dataframes.
````

````python
# Left join
pd.merge(
    left = df_grades,
    right = other_grades,
    on = "names",
    how = "left",
    sort = True,
)

#Podemos ver que haciendo un left join, se unen los dos dataframes, y se agregan las notas faltantes (Science Score y History Score), y el tama침o
#del dataframe resultante es igual al del dataframe de la izquierda, puesto que posee menos datos.
````

````python
#right join
pd.merge(
    left=df_grades,
    right=other_grades,
    on="names",
    how="right"
)

#Podemos ver que haciendo un right join, se unen los dos dataframes, y se agregan las columnas para df_grades que no estaban en other_grades,
#pero si observamos el tama침o del dataframe notamos que pueden haber muchos valores repetidos, puesto que el dataframe de la derecha posee mas datos.
````

````python
#inner join
pd.merge(
    left=df_grades,
    right=other_grades,
    on="names",
    how="inner",
    sort=True,
)

#Podemos ver que haciendo un inner join, se unen los dos dataframes, y se eliminan los valores nulos, pero el tama침o del dataframe resultante es igual al 
#del dataframe de la izquierda, puesto que posee menos datos que el de la derecha, por ende, se eliminan los datos que no estaban en el dataframe de la izquierda.
#En conclusion solo se agregan las notas faltantes al primer dataframe
````

````python
#Creacion data frame final
data_frame_final = pd.merge(
    left=df_grades,
    right=other_grades,
    on="names",
    how="inner",
    sort=True,
)
data_frame_final.head(7)
````
